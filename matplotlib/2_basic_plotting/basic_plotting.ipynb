# Basic Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [4,6,7]

plt.title('Basic Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.plot(x,y, color='green', marker='D', markersize=10)
```

```python
plt.plot(x,y, color='green', linewidth = 2, marker='*', markerfacecolor='green', markersize=10, alpha=2)
```

## Axes labels, Legend, Grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
# Plotting two or more lines on same plot
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

# set axes labels and chart title
plt.title('Weather')
plt.xlabel('Day')
plt.ylabel('Temperature')

# show a legend on the plot
plt.legend(loc='best',shadow=True,fontsize='large')
#plt.grid(True)
```

## Subplot

plt.subplot(nrows,ncols, plot_number)

```python
plt.subplot(1,3,1)
plt.plot(days, max_t, label="max")
# set axes labels and chart title
plt.title('Weather')
plt.xlabel('Day')
plt.ylabel('Temperature')


plt.subplot(1,3,2)
plt.plot(days, min_t, label="min")
# set axes labels and chart title
plt.title('Weather')
plt.xlabel('Day')
plt.ylabel('Temperature')

plt.subplot(1,3,3)
plt.plot(days, avg_t, label="average")
# set axes labels and chart title
plt.title('Weather')
plt.xlabel('Day')
plt.ylabel('Temperature')

fig = plt.figure()
```
