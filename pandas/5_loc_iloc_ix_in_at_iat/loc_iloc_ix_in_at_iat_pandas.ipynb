# .loc, .iloc, .ix, .at, .iat in Pandas

Before applying any Machine Learning Model or Techniques it is necessary to get to know the data attributes and dimensions in order to treat it accordingly.

```python
import pandas as pd
 
df = pd.read_csv(
    filepath_or_buffer='https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
    header=None,
    sep=',')
 
df.columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']
 
df.dropna(how="all", inplace=True) # drops the empty line at file-end
 
df.head()
df.tail()
 
df = df.set_index('class')
```

```python
df.head(5)
```

```python
#df['petal_len'].head()

df.iloc[:3,0:2]
```

```python
df[['petal_len', 'petal_wid']].head(5)
```

**_SELECTING ALL ROWS BY INDEX LABEL_**

```python
# Select all rows with class/index label 'Iris-virginica'
df.loc['Iris-virginica'].head(5)
```

**_SELECTING ROWS IN PANDAS_**

```python
# Select every row up to 5
df.iloc[:4]
 
# Select the forth and fifth row
df.iloc[3:4]
 
# Select every row after the fifth row
df.iloc[4:]
```

**_SELECTING COLUMNS IN PANDAS_**

```python
# Select the first 2 columns
df.iloc[:,:2]
```
